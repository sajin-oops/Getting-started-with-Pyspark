```python
import pyspark
```

```python
from pyspark.sql import SparkSession 
```

```python
spark = SparkSession.builder.appName("dataframe").getOrCreate()
```

```python
df_pyspark = spark.read.csv('employee_dataset.csv',header = True,inferSchema = True)
```

```python
df_pyspark.show()
```

```python
# Filter operations

# salary Less than or equal to 20000
```

```python
df_pyspark.filter('Salary <= 18000').show()
```

```python
df_pyspark.filter('Salary <= 20000').select(['Name','Age']).show()
```

```python
df_pyspark.filter(df_pyspark['Salary']<= 18000).show()
```

```python
df_pyspark.filter((df_pyspark['Salary']<= 20000) & (df_pyspark['Salary'] >= 18000)).show()
```

```python
df_pyspark.filter((df_pyspark['Salary']<= 20000) | (df_pyspark['Salary'] >= 18000)).show()
```

```python
df_pyspark.filter(~(df_pyspark['Salary']<= 29000)).show()
```
